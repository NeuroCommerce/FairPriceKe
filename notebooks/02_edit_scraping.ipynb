```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
import time
import csv
import pandas as pd 
import json

def scrape_product_data(driver, url):
    # Navigate to the page
    driver.get(url)
  
    # Allow page to load
    time.sleep(3)

    # Find all product wrappers
    product_wrappers = driver.find_elements(By.CSS_SELECTOR, ".product-wrapper .thumbnail-wrapper")

    # Create a list to store product data
    products_data = []

    # Loop through each product wrapper and get only the first <a> tag
    for wrapper in product_wrappers:
        try:
            # Find the first <a> tag inside the wrapper
            first_link = wrapper.find_element(By.TAG_NAME, "a")
            product_url = first_link.get_attribute("href")
            print(f"First Product URL: {product_url}")

            # Open product in the same window
            driver.get(product_url)
            time.sleep(2)

            # Extract product information
            try:
                product_name = driver.find_element(By.CLASS_NAME, "product_title").text
            except:
                product_name = "N/A"

            try:
                product_category = driver.find_element(By.CLASS_NAME, "cats-link").text
            except:
                product_category = "N/A"

            try:
                product_price = driver.find_element(By.CLASS_NAME, "price").text
            except:
                product_price = "N/A"

            try:
                short_description = driver.find_element(By.CLASS_NAME, "woocommerce-product-details__short-description").text
            except:
                short_description = "N/A"

            # Add the extracted data to the list
            products_data.append([product_name, product_category, product_price, short_description, product_url])

            # Print the extracted data
            print(f"Product: {product_name}")
            print(f"Category: {product_category}")
            print(f"Price: {product_price}")
            print(f"Short Description: {short_description}")
            print("\n")

            # Go back to the main product list page
            driver.back()
            time.sleep(2)

        except Exception as e:
            print(f"Error processing product: {e}")

    # Save the product data into a CSV file
    with open("product_data.csv", "w", newline="", encoding="utf-8") as csvfile:
        csvwriter = csv.writer(csvfile)
        # Write the header
        csvwriter.writerow(["Product Name", "Category", "Price", "Short Description", "Product URL"])
        # Write the product data
        csvwriter.writerows(products_data)

    print("Data saved to product_data.csv")



```

```python
# Set up Chrome options for headless mode
chrome_options = Options()
chrome_options.add_argument("--headless")  # Run in headless mode
chrome_options.add_argument("--no-sandbox")  # Bypass OS security model
chrome_options.add_argument("--disable-dev-shm-usage")  # Overcome limited resource problems

# Initialize the WebDriver with Chrome options
driver = webdriver.Chrome(options=chrome_options)

url = "https://www.phoneplacekenya.com/product-brand/oppo/"  # Replace with the actual URL
scrape_product_data(driver, url)

# Close the driver
driver.quit()
```
